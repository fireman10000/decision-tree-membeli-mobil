```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("membeli-mobil.xlsx")
```

```python
data.head()
```

```python
inputs = data.drop(['no', 'keputusan_membeli'], axis = 'columns')
target = data['keputusan_membeli']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_jenis_usia = LabelEncoder()
le_pendapatan = LabelEncoder()
le_bekerja = LabelEncoder()
le_peringkat_asesmen_kredit = LabelEncoder()
```

```python
inputs['numerik_jenis_usia'] = le_jenis_usia.fit_transform(inputs['jenis_usia'])
inputs['numerik_pendapatan'] = le_pendapatan.fit_transform(inputs['pendapatan'])
inputs['numerik_bekerja'] = le_bekerja.fit_transform(inputs['bekerja'])
inputs['numerik_peringkat_asesmen_kredit'] = le_peringkat_asesmen_kredit.fit_transform(inputs['peringkat_asesmen_kredit'])
```

```python
inputs.head()
```

```python
inputs_numerik = inputs.drop(['jenis_usia', 'pendapatan', 'bekerja', 'peringkat_asesmen_kredit'], axis = 'columns')
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
decisionTreeModel = model.fit(inputs_numerik, target)
```

```python
#Untuk menginput prediksi data mengikuti model
#([['numerik_jenis_usia', 'numerik_pendapatan', 'numerik_bekerja', 'numerik_peringkat_asesmen_kredit']])
decisionTreeModel.predict([[2,2,2,1]])
```

```python
fn=['numerik_jenis_usia', 'numerik_pendapatan', 'numerik_bekerja', 'numerik_peringkat_asesmen_kredit']
cn=['ya', 'tidak']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=1000)
tree.plot_tree(decisionTreeModel, feature_names = fn, class_names=cn, filled = True);
```
